```python
l1=[]
l2=[]
for i in cacategory['items']:

    key= f'{i['id']}'
    value1 = i['snippet']['title']
    value2 = i['snippet']['channelId']

    l1.append(int(key))
    l2.append([value1, value2])

data = dict(zip(l1,l2))
data.items()
```

```python
rows = [(key, *value) for key, value in data.items()]
columns=['category_id', 'Category_name', 'channel_id']
cacategory_json = pd.DataFrame(rows, columns=columns)
cacategory_json
```

```python
ca_merge = pd.merge(cavideos, cacategory_json, on='category_id', how='left')
ca_merge
```
